```python
import pandas as pd
import matplotlib.pyplot as plt
from src.utils import carregar_dados, faturamento_agrupado, produto_mais_vendido, faturamento_temporal

# Carregar os dados
df = carregar_dados('../data/vendas_detalhadas.csv')

# Visualizar as 5 primeiras linhas
df.head()
```

```python
# Faturamento por cidade
faturamento_cidade = faturamento_agrupado(df, por='Cidade')
faturamento_cidade.plot(kind='bar', title='Faturamento por Cidade', figsize=(10, 5))
plt.ylabel('R$')
plt.xlabel('Cidade')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Faturamento por categoria
faturamento_categoria = faturamento_agrupado(df, por='Categoria')
faturamento_categoria.plot(kind='pie', autopct='%1.1f%%', title='Faturamento por Categoria', figsize=(6, 6))
plt.ylabel('')
plt.show()
```

```python
# Produto mais vendido em quantidade
mais_vendidos = produto_mais_vendido(df)
mais_vendidos.plot(kind='bar', title='Produtos Mais Vendidos (Qtd)', figsize=(10, 5))
plt.ylabel('Quantidade')
plt.xlabel('Produto')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Evolução do faturamento mensal
faturamento_mes = df.resample('M', on='Data')['Faturamento'].sum()
faturamento_mes.plot(kind='line', marker='o', title='Faturamento Mensal', figsize=(12, 5))
plt.ylabel('R$')
plt.xlabel('Mês')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Evolução do faturamento semanal

faturamento_semana = df.resample('W', on='Data')['Faturamento'].sum()
faturamento_semana.plot(kind='line', marker='o', title='Faturamento Semanal', figsize=(12, 5))
plt.ylabel('R$')
plt.xlabel('Semana')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Evolução do faturamento diario
faturamento_dia = faturamento_temporal(df, freq='D')
faturamento_dia.plot(kind='line', marker='o', title='Faturamento ao Longo do Tempo', figsize=(12, 5))
plt.ylabel('R$')
plt.xlabel('Data')
plt.grid(True)
plt.tight_layout()
plt.show()
```
